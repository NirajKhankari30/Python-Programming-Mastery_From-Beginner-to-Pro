```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Step 1: Fetch webpage content with headers
url = "https://webscraper.io/test-sites/e-commerce/allinone/computers/laptops"
headers = {"User-Agent": "Mozilla/5.0"}
req = requests.get(url, headers=headers)

# Step 2: Parse with BeautifulSoup
soup = BeautifulSoup(req.text, "html.parser")

# Step 3: Find all product containers
products = soup.select("div.thumbnail")

# Step 4: Extract data
titles = []
descriptions = []
prices = []
reviews = []

for product in products:
    title = product.select_one("a.title")
    description = product.select_one("p.description")
    price = product.select_one("h4.price")
    review = product.select_one("div.ratings > p.pull-right")

    titles.append(title.text.strip() if title else "N/A")
    descriptions.append(description.text.strip() if description else "N/A")
    prices.append(price.text.strip() if price else "N/A")
    reviews.append(review.text.strip() if review else "N/A")

# Step 5: Create and display DataFrame
df = pd.DataFrame({
    "Titles": titles,
    "Descriptions": descriptions,
    "Prices": prices,
    "Reviews": reviews
})

df.head()
```
